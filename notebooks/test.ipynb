```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x44 = np.linspace(-5.0,5.0,50)
y44 = np.tanh(np.exp(x44)) * ((x44 > -1) * x44)
z44 = x44 * np.tanh(np.log(1+np.exp(x44)))
```

```python
fig = plt.figure(figsize=(30,30))
ax = fig.add_subplot(111)
plt.plot(x44,y44, 'r', label= "SharkFin", linewidth=5.0)
plt.plot(x44,z44, 'b*-', label= "Mish", linewidth=2.0)
ax.legend(loc='best', fontsize=35)
ax.grid(linestyle = ':')
ax.tick_params(axis='both', which='major', labelsize=35)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
plt.savefig('graph_te_relu.png', bbox_inches='tight')
plt.show()
```

```python
x=np.linspace(-3.0,3.0,50)
x1=np.linspace(-3,3,50)
x2=np.linspace(-10,10,50)
beta1=x*np.tanh(np.log(1+np.exp(x)))
beta=np.tanh(np.exp(x)) * ((x > -1) * x)

import scipy.misc
def mish(x1):
    return x1*np.tanh(np.log(1+np.exp(x1)))
def aria(x):
    return np.tanh(np.exp(x)) * ((x > -1) * x)
dev4=scipy.misc.derivative(aria, x1, n=1)
dev5=scipy.misc.derivative(aria, x1, n=2)
dev6=scipy.misc.derivative(aria, x2, n=1)
dev1= scipy.misc.derivative(mish, x1, n=1)
dev2=scipy.misc.derivative(mish, x1, n=2)
dev3=scipy.misc.derivative(mish, x2, n=1)
```

```python
fig7 = plt.figure(figsize=(30,10))
ax2 = fig7.add_subplot(131)
plt.plot(x,beta1, 'r*-', label= "Mish")
plt.plot(x,beta, 'ko-', label= "SharkFin ( = -1.0)")
plt.grid()
ax2.legend(loc='best', fontsize=20)
ax2.set_title('Function Curves of SharkFin ( = -1.0)', fontdict={'fontsize': 20, 'fontweight': 'medium'})
ax2.tick_params(axis='both', which='major', labelsize=20)
ax2.tick_params(axis='both', which='minor', labelsize=20)
plt.axhline(linewidth=1, color='k') #adds thick red line @ y=0
plt.axvline(linewidth=1, color='k') #adds thick red line @ x=0
ax = fig7.add_subplot(132)
plt.plot(x1,dev1,'m',label="1st Derivative of Mish")
plt.plot(x1,dev2,'mo-', label="2nd Derivative of Mish")
plt.plot(x1,dev4,'k',label='1st Deriavtaive of SharkFin ( = -1.0)')
plt.plot(x1,dev5,'ko-',label='2nd Deriavtaive of SharkFin ( = -1.0)')
ax.set_title('Derivative of SharkFin ( = -1.0)', fontdict={'fontsize': 20, 'fontweight': 'medium'})
ax.tick_params(axis='both', which='major', labelsize=15)
ax.tick_params(axis='both', which='minor', labelsize=15)
plt.axhline(linewidth=1, color='r') #adds thick red line @ y=0
plt.axvline(linewidth=1, color='r') #adds thick red line @ x=0
plt.grid()
ax.legend(loc='best', fontsize=15)
ax1=fig7.add_subplot(133)
plt.plot(x2,dev3,'m',label="1st Derivative of Mish ")
plt.plot(x2,dev6,'k',label='1st Derivative of  SharkFin \n ( = -1.0)')
ax1.set_title('Derivative of SharkFin ( = -1.0) (Increased Range)', fontdict={'fontsize': 20, 'fontweight': 'medium'})
ax1.tick_params(axis='both', which='major', labelsize=15)
ax1.tick_params(axis='both', which='minor', labelsize=15)
plt.axhline(linewidth=1, color='r') #adds thick red line @ y=0
plt.axvline(linewidth=1, color='r') #adds thick red line @ x=0
plt.grid()
ax1.legend(loc=4, fontsize=15)
plt.savefig('All.png', bbox_inches='tight')
plt.show()
```
